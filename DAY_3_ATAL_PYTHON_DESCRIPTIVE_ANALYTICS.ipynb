```python
plot_master = pd.read_csv(filePath+"/DAY_5_Ground Mounted_Data_Data_Set.csv")

block=[]

total=1
#  creating blocks 4 block == 1 hrs 
for i in range(1, 97,4): # for loop range(low, high, steps)
    
    block = block+[f'Block {total}']*4
        
    total+=1

plot_master['TIME_BLOCK']=block

plot_master[['TIME', 'TIME_BLOCK']]

pie_plot=plot_master[['TIME_BLOCK','Generation in each block']].groupby(['TIME_BLOCK'],sort=False).sum()

# plot_master
plt.figure(figsize=(18,8))
plt.pie(pie_plot, autopct="%0.0f%%", labels=pie_plot.index)
plt.title("ENERGY GENERATION PER BLOCK")
plt.tight_layout()
```
